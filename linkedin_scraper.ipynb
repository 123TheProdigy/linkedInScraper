```python
def searchDF(df, keyword):
    titles = list(df['currentTitle'])
    urls = list(df['profileURL'])
    narrowS = []
    for i in range(0, len(titles)):
        if keyword in titles[i].lower():
            conc = [titles[i], urls[i]]
            narrowS.append(conc)
    newDF = pd.DataFrame(narrowS, columns = ['title', 'link'])
    return newDF
```

```python
def next_page(link):
    names = []
    rev = []
    
    driver2 = webdriver.Chrome(r"C:\Users\Allen\chromedriver")
    driver2.get(link)
    selenium_html = driver2.page_source
    soup = BeautifulSoup(selenium_html, 'lxml')
    
    find_next = soup.find('span', {"class" : "prev-next-paging__button-text"})
    next_text = BeautifulSoup(str(find_next), features = "html.parser").text

    while(next_text == 'Next'):
        print("entered")
        review_name_specific = soup.find_all("a", {"class" : "unstyled bold articleLink"})
        actual_review = soup.find_all("div", {"class" : "the_review"})
        
        for person in review_name_specific:
            text = BeautifulSoup(str(person), features = "html.parser").text
            names.append(text)
        for review in actual_review:
            text = BeautifulSoup(str(review), features = "html.parser").text
            rev.append(text)

        numReviews = len(rev)
        for item in range(0, numReviews):
            rev[item] = re.sub(" +", " ", rev[item])
            rev[item] = re.sub("\r\n", "", rev[item])
        
        print(names)
        print(rev)
            
        driver2.find_element("xpath", "//span[text()='Next']").click()
        selenium_html = driver2.page_source
        soup = BeautifulSoup(selenium_html, 'lxml')
        
        find_next = soup.find('span', {"class" : "prev-next-paging__button-text"})
        next_text = BeautifulSoup(str(find_next), features = "html.parser").text

    return (np.stack((names, rev), axis = -1))


## run try/except
        
```
